# Mengimport file csv
Sebelum membaca file csv, kita haru mengimport library pandas, sklearn, dan numpy

Lalu menentukan untuk heading tabel sebagai FEATURE_NAMES

Lalu menentukan variabel iris sebagai mengimport data csv

Membuat kerangka tabel dengan variabel X dan isi dengan kolom variabel FEATURE_NAMES

Mengubah kolom variety menjadi data int

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

FEATURE_NAMES = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

iris = pd.read_csv('irisdata.csv')
X = pd.DataFrame(iris, columns = FEATURE_NAMES)

target = preprocessing.LabelEncoder()
target.fit(iris.variety)
y = target.transform(iris.variety)
```

# Menampilkan 150 data pada variabel X

```python
X[:150]
```

# Menampilkan 150 data pada variabel y

```python
y[:150]
```

# Membuat data training dan data test dengan test size 35% serta random state dengan NIM = H96218062 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.35, random_state = 96218062)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape, y_test.shape)
```

# Mengimport library untuk Decision Tree dan menentukan kriteria entropy dan gini dengan kedalaman 3,4,5,7

```python
from sklearn.tree import DecisionTreeClassifier

#entropy
model3 = DecisionTreeClassifier(criterion="entropy", max_depth = 3)
model3.fit(X_train,y_train)

model4 = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
model4.fit(X_train,y_train)

model5 = DecisionTreeClassifier(criterion="entropy", max_depth = 5)
model5.fit(X_train,y_train)

model7 = DecisionTreeClassifier(criterion="entropy", max_depth = 7)
model7.fit(X_train,y_train)

#gini
mdl3 = DecisionTreeClassifier(criterion="gini", max_depth = 3)
mdl3.fit(X_train,y_train)

mdl4 = DecisionTreeClassifier(criterion="gini", max_depth = 4)
mdl4.fit(X_train,y_train)

mdl5 = DecisionTreeClassifier(criterion="gini", max_depth = 5)
mdl5.fit(X_train,y_train)

mdl7 = DecisionTreeClassifier(criterion="gini", max_depth = 7)
mdl7.fit(X_train,y_train)
```
